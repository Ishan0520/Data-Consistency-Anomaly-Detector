```python
import pandas as pd
import numpy as np

# Upload CSV File
from google.colab import files
uploaded = files.upload()  # Upload your dataset manually

# Load the dataset
filename = list(uploaded.keys())[0]  # Get the uploaded file name
df = pd.read_csv(filename, encoding='latin-1')  # Read CSV into DataFrame

# Function to find the primary key
def find_primary_key(df):
    for column in df.columns:
        if df[column].is_unique:
            print(f"‚úÖ '{column}' can be a Primary Key")
        else:
            print(f"‚ùå '{column}' is NOT a Primary Key")

# Run the primary key check
find_primary_key(df)

```

```python
def data_quality_report(df):
    print("\nüìä Data Quality Report:")

    # 1Ô∏è‚É£ Check for missing values
    missing_values = df.isnull().sum()
    print("\nüõë Missing Values:")
    print(missing_values[missing_values > 0])

    # 2Ô∏è‚É£ Check for duplicate rows
    duplicates = df.duplicated().sum()
    print("\nüîÅ Duplicate Rows:", duplicates)

    # 3Ô∏è‚É£ Check data types
    print("\nüìå Column Data Types:")
    print(df.dtypes)

# Run Data Quality Check
data_quality_report(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizing Missing Values
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()
```
